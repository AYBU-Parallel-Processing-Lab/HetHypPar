```python
path_weight_dir = Path("data/weights")
path_weights = list(path_weight_dir.glob("*.txt"))
```

```python
def parse_weight(wpath: Path):
  wlist_str = wpath.read_text().split()
  return [
    float(wstr) for wstr in wlist_str
  ]
parse_weight(path_weights[5])
```

```python
path_weight = path_weights[0]
list_weight = parse_weight(path_weight)
num_parts = len(list_weight)
num_parts
```

```python
cmd_patpart = "patpart {matrix_path} {npart} {weight_path} {imbal_percent} {seed} {output_path} {log_path}"
imbal = 0.10
imbal_percent = int(imbal*100)
seed = 42

def format_cmd(matrix_name : str, path_rank_weight : Path):
  nparts = len(parse_weight(path_rank_weight))
  
  path_data_dir : Path = Path("./data/matrices") / matrix_name
  path_in_dir = path_data_dir / "in"
  path_out_dir = path_data_dir / "out"
  
  path_part_dir = path_in_dir / "part"
  path_part_dir.mkdir(parents=True, exist_ok=True)
  
  outname = f"{path_rank_weight.stem}_i{imbal_percent}"
  
  path_output = path_part_dir / f"{outname}.part"
  path_log = path_part_dir / f"{outname}.log"
  
  return cmd_patpart.format(
    matrix_path=f"/matrices/{matrix_name}.mtx",
    npart=nparts,
    weight_path=path_rank_weight,
    imbal_percent=imbal_percent,
    seed=seed,
    output_path = path_output,
    log_path = path_log
  ) 
  
```

```python
my_cmd = format_cmd("cage13", path_weights[0])
my_cmd
```

```python
import os
```

```python
os.system(my_cmd)
```

# Get matrices

```python
from scipy.io import mminfo
```

```python
# path_mtx = mtx_files[0]

def check_mtx(path_mtx: Path):
  numrows, numcols, nnz, format_type, field, symmetry = mminfo(path_mtx)
  if numrows != numcols:
    # print(f"Skipping {matname}: not square ({numrows}x{numcols})")
    return False
  
  if symmetry == 'symmetric':
      # print(f"Skipping {matname}: symmetric matrix")
      return False
    
  return True
  
```

```python
matrices_dir = Path("/matrices")
```

```python
mtx_files = sorted(matrices_dir.glob("*.mtx"))
nbefore = len(mtx_files)
mtx_files = [
  path_mtx for path_mtx in mtx_files if check_mtx(path_mtx)
]
nafter = len(mtx_files)
print(f"Filtered {nbefore - nafter} out of {nbefore} matrices")
```

```python
[
  format_cmd(mtx_file.stem, path_weight)
  for mtx_file in mtx_files
  for path_weight in path_weights
]
```

```python
path_data_dir : Path = Path("./data/matrices")
```

```python
def prep_mat_dir(matnam : str):
  path_basedir = path_data_dir / matnam
  path_indir = path_basedir / "in"
  path_indir.mkdir(parents=True, exist_ok=True)
  path_outdir = path_basedir / "out"
  path_outdir.mkdir(parents=True, exist_ok=True)
  
```

```python
for mtx_file in mtx_files:
  prep_mat_dir(mtx_file.stem)
  # break
```

```python
from tqdm.notebook import tqdm
import os
```

```python
for mtx_file in tqdm(mtx_files):
  for path_weight in path_weights:
    my_cmd = format_cmd(mtx_file.stem, path_weight)
    os.system(my_cmd)
    # break
  # break
```
